```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image
from tensorflow.keras import layers, models
```

```python
(x_train, y_train) , (x_test, y_test) = tf.keras.datasets.mnist.load_data()

x_train, x_test = x_train / 255.0 , x_test / 255.0
```

```python
i = 1
plt.imshow(x_test[i], cmap='plasma')
plt.show()
```

```python
model = models.Sequential([
    layers.Flatten(input_shape=(28,28)),
    layers.Dense(128, activation='relu'),
    layers.Dense(10, activation='softmax')
])
```

```python
# Definição do modelo
model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)
```

```python
#Treino
model.fit(
    x_train,
    y_train,
    epochs=5,
    batch_size=32,
    verbose=1
)
```

```python
#teste da acuracia
loss, acuracia = model.evaluate(x_test, y_test)
print(f'Acurácia: {acuracia}')
```

```python
# Teste com as proprias imagens do dataset
i = 14
img = x_test[i].reshape(1,28,28)

pred = model.predict(img).argmax()

plt.imshow(x_test[i], cmap='plasma')
plt.title(f"Real: {y_test[i]} | Previsto: {pred}")
plt.show()
```
